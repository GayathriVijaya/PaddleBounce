```python
from turtle import *
import turtle
import winsound

 


wn=turtle.Screen()
wn.title("PING-PONG GAME")
wn.bgcolor("plum")
wn.setup(width=800,height=600)
wn.tracer(0) #stops the window from updating so we need to update the window manually


#score
score_a=0
score_b=0


#object1
paddle1=turtle.Turtle() #Turtle is class name , turtle modules name
paddle1.speed(0) #speed of animation sets the speed to the max speed
paddle1.shape("square")
paddle1.color("slate blue")
paddle1.shapesize(stretch_wid=5,stretch_len=1) #to increase the size of the paddle
paddle1.penup() #o drawing when moving  , turtle ussually draw line we dont need it
paddle1.goto(-350,0)  #initialize the positio of the obj

#object2
paddle2=turtle.Turtle() #Turtle is class name , turtle modules name
paddle2.speed(0) #speed of animation sets the speed to the max speed
paddle2.shape("square")
paddle2.color("slate blue")
paddle2.shapesize(stretch_wid=5,stretch_len=1) #to increase the size of the paddle
paddle2.penup() #no drawing when moving  , turtle ussually draw line we dont need it
paddle2.goto(350,0)

#ball
ball=turtle.Turtle() #Turtle is class name , turtle modules name
ball.speed(0) #speed of animation sets the speed to the max speed
ball.shape("circle")
ball.color("teal")
ball.penup() #o drawing when moving  , turtle ussually draw line we dont need it
ball.goto(0,0)
ball.dx = 2#every time the ball moves it moes by 2 pixels alonf positive x and y
ball.dy = 2


#pen
pen=turtle.Turtle()
pen.speed(0)
pen.color("white")
pen.penup()
pen.hideturtle()# we dont wanna see it so only the text is needed
pen.goto(0,260)  #height is 300
pen.write("Player A: 0  Player B: 0",align="center",font=("Arial",24,"normal"))



#function
def paddle1_up():
    y= paddle1.ycor()#a fuction in turtle module that gies the y coordinate
    y+=20
    paddle1.sety(y)
    
def paddle1_down():
    y= paddle1.ycor()#a fuction in turtle module that gies the y coordinate
    y-=20
    paddle1.sety(y)
    
def paddle2_up():
    y= paddle2.ycor()#a fuction in turtle module that gies the y coordinate
    y+=20
    paddle2.sety(y)
    
def paddle2_down():
    y= paddle2.ycor()#a fuction in turtle module that gies the y coordinate
    y-=20
    paddle2.sety(y)
    
#the function need to be called by keyboard
#keyboard binding
wn.listen()  #listen to the keyboard input  wn-window 
wn.onkeypress(paddle1_up,"u") #when the input from keyboard is w then call function paddle1_up
wn.onkeypress(paddle1_down,"d")
wn.onkeypress(paddle2_up,"Up") 
wn.onkeypress(paddle2_down,"Down")

    

#main game loop
while True:
    wn.update()#updates the screen everytime the loop runs
    #move ball
    ball.setx(ball.xcor() + (ball.dx/32))  #xcor is the default pos of x axis that is 0 and  is 2 added evvery time the loop is done
    ball.sety(ball.ycor() + (ball.dy/32))  #by 32 to decrease the speed of the ball
    
    
    #border checking
    if ball.ycor()>290:  #if the ball is about to touch the wall thenat is 600 is height 300 along +y axis 
        ball.sety(290) #  and -300 along -y axis so if it is >290 then it should trace back
        ball.dy *= -1
       
    if ball.ycor()<-290:  
        ball.sety(-290) 
        ball.dy *=-1 
        
    if ball.xcor()>390: 
        ball.goto(0,0) 
        ball.dx *=-1 
        score_a+=1
        pen.clear()
        pen.write("Player A: {}   Player B: {}".format(score_a,score_b),align="center",font=("Arial",24,"normal"))
       
    if ball.xcor()<-390: 
        ball.goto(0,0) 
        ball.dx *=-1
        score_b+=1
        pen.clear()
        pen.write("Player A: {}   Player B: {}".format(score_a,score_b),align="center",font=("Arial",24,"normal"))
        
      #paddle and ball collission
    if (ball.xcor() >340 and ball.xcor()<350) and (ball.ycor() < paddle2.ycor() + 40  and ball.ycor()> paddle2.ycor() - 40):
        ball.setx(340)
        ball.dx*=-1
       
    if (ball.xcor() <-340 and ball.xcor()>-350) and (ball.ycor() < paddle1.ycor() + 40  and ball.ycor()> paddle1.ycor() - 40):
        ball.setx(-340)
        ball.dx*=-1
      
```
